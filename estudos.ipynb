# Funções de Ativação

```python
import numpy as np
import matplotlib.pyplot as plt

z = np.linspace(-2, 2, 100)

def sigmoid(z):
    return 1 / (1 + np.exp(-z))

def softmax(z):
    e_z = np.exp(z)
    return e_z / e_z.sum()

def relu(z):
    return np.maximum(0, z)

sigmoid_values = sigmoid(z)
softmax_values = softmax(z)
relu_values = relu(z)

plt.figure(figsize=(16, 6))
plt.plot(z, sigmoid_values, label="Sigmóide", color="blue")
plt.title("Função Sigmóide")
plt.xlabel("z")
plt.ylabel("x_sigmoide")
plt.grid(True)
plt.show()

plt.figure(figsize=(16, 6))
plt.plot(z, softmax_values, label="Softmax", color="red")
plt.title("Função Softmax")
plt.xlabel("z")
plt.ylabel("s_softmax")
plt.grid(True)
plt.show()

plt.figure(figsize=(16, 6))
plt.plot(z, relu_values, label="ReLU", color="green")
plt.title("Função ReLU")
plt.xlabel("z")
plt.ylabel("x_relu")
plt.grid(True)
plt.show()
```

# Bias

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x, scale, shift=0):
    return 1 / (1 + np.exp(-scale * (x + shift)))

x = np.linspace(-10, 10, 400)

y1_orig = sigmoid(x, 0.5)
y2_orig = sigmoid(x, 1.0)
y3_orig = sigmoid(x, 2.0)

y1_new = sigmoid(x, 1.0, -5*1.0)
y2_new = sigmoid(x, 1.0, 0*1.0)
y3_new = sigmoid(x, 1.0, 5*1.0)

plt.style.use('default')

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Originais
axes[0].plot(x, y1_orig, label='sig(0.5*x)', color='red')
axes[0].plot(x, y2_orig, label='sig(1.0*x)', color='green')
axes[0].plot(x, y3_orig, label='sig(2.0*x)', color='blue')
axes[0].legend()
axes[0].set_xlabel('Input')
axes[0].set_ylabel('Output')
axes[0].set_title('Exemplo de Sigmóide sem viés')

# Viéses
axes[1].plot(x, y1_new, label='sig(1.0*x + -5*1.0)', color='red')
axes[1].plot(x, y2_new, label='sig(1.0*x + 0*1.0)', color='green')
axes[1].plot(x, y3_new, label='sig(1.0*x + 5*1.0)', color='blue')
axes[1].legend()
axes[1].set_xlabel('Input')
axes[1].set_ylabel('Output')
axes[1].set_title('Exemplo de Sigmóide com viés')

plt.tight_layout()
plt.show()
```

# Overfitting

```python
import matplotlib.pyplot as plt
import numpy as np

train_sizes = np.linspace(0.1, 1.0, 25)
train_error_adequado = 1 - train_sizes + np.random.normal(0, 0.2, 25)
val_error_adequado = 1.2 - train_sizes + np.random.normal(0, 0.2, 25)

train_error_overfitting = 1 - train_sizes + np.random.normal(0, 0.2, 25)
val_error_overfitting = 3 - train_sizes + np.random.normal(0, 0.5, 25)

plt.figure(figsize=(16, 6))

# Gráfico 1: Modelo Adequado
plt.subplot(1, 2, 1)
plt.plot(train_sizes, train_error_adequado, color='g', label='Erro de Treinamento')
plt.plot(train_sizes, val_error_adequado, color='r', label='Erro de Validação')
plt.fill_between(train_sizes, train_error_adequado - 0.2, train_error_adequado + 0.2, color='g', alpha=0.1)
plt.fill_between(train_sizes, val_error_adequado - 0.2, val_error_adequado + 0.2, color='r', alpha=0.1)
plt.title('Curva de aprendizado em modelo adequado')
plt.xlabel('Iteração')
plt.ylabel('Erro Quadrático Médio (MSE)')
plt.legend(loc='best')
plt.grid(True)

# Gráfico 2: Modelo com Sobreajuste
plt.subplot(1, 2, 2)
plt.plot(train_sizes, train_error_overfitting, color='g', label='Erro de Treinamento')
plt.plot(train_sizes, val_error_overfitting, color='r', label='Erro de Validação')
plt.fill_between(train_sizes, train_error_overfitting - 0.2, train_error_overfitting + 0.2, color='g', alpha=0.1)
plt.fill_between(train_sizes, val_error_overfitting - 0.2, val_error_overfitting + 0.2, color='r', alpha=0.1)
plt.title('Curva de aprendizado em modelo com overfitting')
plt.xlabel('Iteração')
plt.ylabel('Erro Quadrático Médio (MSE)')
plt.legend(loc='best')
plt.grid(True)

plt.tight_layout()
plt.show()
```
